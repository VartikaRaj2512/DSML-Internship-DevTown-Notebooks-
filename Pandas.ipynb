```python
import re
inp = input("Enter any string: ")
def check_me():
  patterns = 'ab*'
  if re.search(patterns, inp):
    print("yes, matched")
  else:
    return('Not matched!')

check_me()
```

```python
def pat():
#str=input("enter any string!")
  stri=input("enter any string!:")
  pattern="ab*?"
  if re.findall(pattern,stri):
    print("matched")
  else:
    print("no match")
```

```python
def email_validator(mail):
  if len(mail)<=100:
    valid=re.match("[\w]+@[a-z]{3,6}[.][a-z]{3,5}$",mail)
    if valid:
      print("Valid email")
    else:
      print("Invalid email")
mail=input("Enter email: ")
email_validator(mail)
```

```python
rd = "0--1--1-a9-b9-c9-d9-e9-f9-g9-h9-i9-j9"
d=re.findall('\d',rd)
o=''
count=0
for i in d:
  if count==3:
    o+='-'
  else:
    o+=i
  count+=1
print(o)
```
