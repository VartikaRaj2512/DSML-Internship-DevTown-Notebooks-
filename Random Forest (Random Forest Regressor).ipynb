# Import libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
```

# Importing the Dataset

Link: https://raw.githubusercontent.com/mk-gurucharan/Regression/master/IceCreamData.csv

```python
df = pd.read_csv('https://raw.githubusercontent.com/mk-gurucharan/Regression/master/IceCreamData.csv')
```

# Data Analysis (EDA, DM, DE, DC, DV)

```python
df.head()
```

```python
df.info()
```

# Assigning features and target

```python
X = np.array(df.Temperature)
y = np.array(df.Revenue)
```

# Splitting the dataset

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.05)
```

# Model Selection

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
regressor = RandomForestRegressor(n_estimators=10, random_state=0)
```

# Training the model

```python
regressor.fit(X_train.reshape(-1,1), y_train.reshape(-1,1))
```

# Testing the model

```python
y_pred = regressor.predict(X_test.reshape(-1,1))
```

```python
pred = pd.DataFrame({"Actual":y_test.reshape(-1), 
                     "Prediction":y_pred.reshape(-1)})
```

```python
pred.head()
```

```python
plt.scatter(X_test,y_test, color='red')
plt.scatter(X_test,y_pred, color='green')
plt.xlabel("Temperature")
plt.ylabel("Revenue")
```

```python
sns.heatmap(pred.corr(), annot=True, cmap='Greens')
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, y_pred)
```

# Splitting the data

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2,
                                                    random_state=0)
```

```python
X_test
```

# Model Selection

```python
regressor = RandomForestRegressor()
```

# Hyper Parameter Tuning

```python
n_estimators = [int(i) for i in np.linspace(start=100, stop=1200, num=12)]

max_features = ['auto','sqrt']

max_depth = [int(i) for i in np.linspace(start=5, stop=30, num=6)]

min_samples_split = [2, 5, 10, 15, 100]

min_samples_leaf = [1, 2, 5,10]
```

```python
random_grid = {'n_estimators':n_estimators,
               'max_features':max_features,
               'max_depth':max_depth,
               'min_samples_split':min_samples_split,
               'min_samples_leaf':min_samples_leaf}
print(random_grid)
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
rf_regressor = RandomizedSearchCV(estimator=regressor,
                                  param_distributions=random_grid,
                                  scoring='neg_mean_squared_error',
                                  cv=5,
                                  verbose = 2,
                                  random_state=42,
                                  n_jobs=1)
```

# Training the model

```python
rf_regressor.fit(X_train, y_train)
```

```python
rf_regressor.best_params_
```

# Testing the model

```python
y_pred = rf_regressor.predict(X_test)
```

```python
plt.scatter(y_test, y_pred)
```

```python
finaldf = pd.DataFrame({"Actual":y_test, "Predicted":y_pred})
```
