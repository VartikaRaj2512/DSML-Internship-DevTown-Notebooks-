```python
class Addition:
  f = 0
  s = 0
  a = 0

  # parameterized constructor
  def __init__(self, f1, s1):
    self.f = f1
    self.s = s1

  def display(self):
    print("First number = " + str(self.f))
    print("Second number = " + str(self.s))
    print("Addition of two numbers = " + str(self.a))

  def calculate(self):
    self.a = self.f + self.s
```

```python
x=eval(input("Enter number:"))
y=eval(input("Enter number:"))
```

```python
obj = Addition(x,y)
```

```python
obj.calculate()
```

```python
obj.display()
```

```python
class Icecream:

  def __init__ (self, flavour,color,structure):
    self.flavour = flavour
    self.color = color
    self.structure = structure

  def eat(self):
    print(f"The {self.flavour} has the color {self.color}.")

  def eat1(self):
    print(f"The {self.flavour} has the structure {self.structure}!!")

```

```python
obj = Icecream("Strawberry","Pink","Cone")
```

```python
obj.eat()
```

```python
class P:
  a = 1
  b = 1
  answer = 1
  def __init__(self,a,b):
    self.a = a
    self.b = b
  def print_numbers(self):
    print("Your first number is:",self.a)
    print("Your second number is:",self.b)
    print("Multiplied value of two numbers:",self.answer)
  def multiplication(self):
    self.answer = self.a*self.b
```

```python
obj = P(2,2)
```

```python
obj.multiplication()
```

```python
obj.print_numbers()
```
