# **Python**

Python is an interpreted high-level programming language. High-level language are the languages that are closer to human languages (Human understandable) and further from machine languages.

**Designed By:** Guido van Rossum

[Features of Python](https://cdn.educba.com/academy/wp-content/uploads/2019/10/Python-Features.png.webp)

A programming language need eith the Compiler or an Interpreter to converts the programming language/code into machine language/code so that the Computer can understands it.


**Interpreter:** It translates only one statement of the program at a time. Ex: Python, JavaScript, Ruby

**Compiler:** Scans the entire program and translates it as a whole into machine code. Ex: C, C++, Java

**Drawbacks of Python:** Slower than other prgramming languages such as C, C++.


**Fun Fact:** Guido van Rossum wanted to select a name which is unique, sorted, and little-bit mysterious. So he decided to select naming Python after the popular BBC comedy series "Monty Python's Flying Circus" for their newly created programming language.

# Naming the variable

```python
Rule-1: You should start variable name with an alphabet or underscore(_) character.
Rule-2: A variable name can only contain A-Z,a-z,0-9 and underscore(_).
Rule-3: You cannot start the variable name with a number.
Rule-4: You cannot use special characters with the variable name such as such as $,%,#,&,@.-,^ etc.
Rule-5: Variable names are case sensitive. For example str and Str are two different variables.
Rule-6: Do not use reserve keyword as a variable name for example keywords like class, for, def, del, is, else, try, from, etc. more examples are given below and as we go through the course we will come across many more. Creating names that are descriptive of the values often will help you avoid using any of these words.
```
