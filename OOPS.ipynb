# Inheritance

```python
class father(): # Base/Parent class
  def write(self):
    return "yes"

class son(father): # Derived/Child class
  def write(self):
    return "yes"
```

# Single Inheritance.

```python
class Animal:
  def speak(self):
    print("Animal is speaking")

class Dog(Animal):
  def bark(self):
    print("The dog barks")
  # def speak(self):
  #   print("Cat is speaking")
```

```python
obj = Dog()
```

```python
obj.bark()
```

```python
obj.speak()

# Animal is speaking - Parent class will get preference
# Cat is speaking
```

# Multi Level

```python
class Animal:
  def speak(self):
    print("Animal is speaking")

class Dog(Animal):
  def bark(self):
    print("The dog barks")

class Puppy(Dog):
  def eat(self):
    print("The puppy is eating")
  # def bark(self):
  #   print("The puppy is not barking")
  
```

```python
obj = Puppy()
```

```python
obj.eat()
```

```python
obj.speak()

# Animal is speaking
# error
```

# Multiple Inheritance

```python
class Samp1:
  def new1(self):
    print("This is the 1st class")
  def new2(self):
    print("This is the 1st class")

class Samp2:
  def new1(self):
    print("This is the 2nd class")
  def new2(self):
    print("This is the 2nd class")


class finale(Samp1,Samp2):
  def eat(self):
    print("The dog is eating")
  def new1(self):
    print("This is the child class")
```

```python
obj = finale()
```

```python
obj.new1()

# 1st 2nd
# 2nd 1st
# 1st
# 2nd
```

# Method Overiding

```python
class Animal:
  def speak(self):
    return 10

class Dog(Animal):
  def speak(self):
    return 7

class Cat(Animal):
  def speak(self):
    return 2
```

```python
obj1 = Animal()
obj2 = Dog()
obj3 = Cat()
```

```python
obj2.speak()
```
