# Determinant

```python
# Square matrix -> No. of rows = No. of columns
# Square matrix -> No. of rows & No. of columns = Same

detarray = np.array([[-4,-10],[0,5]])
```

```python
detarray

det = (-4*5) - (-10*0) = -20
```

```python
[a      (-)b    c]
[(-)d   e    (-)f]
[g     (-)h     i]
```

```python
# np.linalg.det(arr)

np.linalg.det(detarray)
```

* around()

```python
arr = np.array([3.501, 0.98])
```

```python
np.around(arr,decimals=3)
# [0,0]
# [0,1]
# [1,1]
# [3.501, 1.000]
# [3.501, 0.98]
# [3.500, 1.000]
```
