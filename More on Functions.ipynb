```python
def isPalindrome(string):
  left_pos = 0
  right_pos = len(string) - 1

  while right_pos >= left_pos:
    if not string[left_pos] == string[right_pos]:
      return False
    left_pos += 1
    right_pos -= 1
    return True
print(isPalindrome('Malayalam'))
```

```python
def test_prime(n):
  if (n==1):
    return False
  elif (n==2):
    return True
  else:
    for x in range(2,n):
      if(n % x==0):
        return False
  return True
print(test_prime(27))
```

```python
def prime(num):
  for i in range(2,num):
    if num%i == 0:
      print(num, "is not prime")
      break

    else:
      print(num, "is prime")
```

```python
prime(9)
```

```python
def check_prime(no):
  ct = 0
  for i in range(1,no+1):
    if no%i == 0:
      ct += 1
      if ct==2:
        print("{} is a prime number.".format(no))
    else:
      print("{} is a composite number".format(no))

check_prime(int(input("Enter a number : ")))
```

```python
def string_test(s):
  d={"UPPER_CASE":0, "LOWER_CASE":0}
  for c in s:
    if c.isupper():
      d["UPPER_CASE"]+=1
    elif c.islower():
      d["LOWER_CASE"]+=1
    else:
      pass
  print ("Original String : ", s)
  print ("No. of Upper case characters : ", d["UPPER_CASE"])
  print ("No. of Lower case Characters : ", d["LOWER_CASE"])
```

```python
string_test("MobiUs")
```

```python
def multiply(numbers):
  total = 1
  for x in numbers:
    total *= x
  return total
  
print(multiply((1,2,3,4)))
```

```python
def multiply(ls):
  product = 1
  for i in ls:
    product = product*i
  return product

print("The product : - {}".format(multiply([int(input()) for i in range(int(input("Enter Number of integers : - ")))])))
```

```python
def unique_list(l):
  x = []
  for a in l:
    if a not in x:
      x.append(a)
  return x

print(unique_list([1,2,3,3,3,3,4,5]))
```
