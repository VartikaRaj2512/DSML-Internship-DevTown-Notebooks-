4. Create DataFrame using Series

```python
data = {"Karan_Marks":pd.Series([50,77,65,98],index=["Test1","Test2","Test3","Test4"]),
        "Arjun_Marks":pd.Series([67,89,32,98,50],index=["Test-1","PTest1","Test-2","PTest2","Test-3"])}
df = pd.DataFrame(data,columns=["A","B"],index=["AA","AB"])
```

```python
df
# If columns headings are changed to A & B and all the values -> NaN
# If only columns headings are changed to A & B but the data remain the same
```
