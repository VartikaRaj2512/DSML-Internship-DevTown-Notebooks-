* Labels

```python
plt.plot(y)
plt.plot(x)

plt.xlabel('Height')
plt.ylabel('Weight')

plt.show()
```

* Title

```python
plt.plot(y)
plt.plot(x)

modification = {'color':'red', 'size':15, 'family':'serif'}

plt.title("Height V/s Weight Graph", fontdict=modification, loc='right')
plt.xlabel('Height')
plt.ylabel('Weight', fontdict=modification)



plt.show()
```

* Grid

```python
plt.plot(y)
plt.plot(x)

plt.xlabel('Height')
plt.ylabel('Weight')

plt.grid()

plt.show()
```

```python
plt.plot(y)
plt.plot(x)

plt.xlabel('Height')
plt.ylabel('Weight')

plt.grid(color='black', ls='--', linewidth=2, axis='x')

plt.show()
```

* Ticks  & Ticks Label

```python
x = [3,8,1,10]
y = [6,2,7,11]

# plt.plot(x,y)

fig = plt.figure()
ax = fig.add_axes([1,1,2,2])

ax.plot(x,y)
ax.set_title("Animal")
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')

# ax.set_xticks([100,102,103,104])
ax.set_xticklabels([100,102,103,104,106,108])
ax.set_yticklabels(['One','Two','Three','Four','Five','Six'])



```
